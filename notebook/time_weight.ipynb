```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

def time_decay_weight(days: np.ndarray, decay_rate: float) -> np.ndarray:
    return np.exp(-decay_rate * days)

days = np.arange(1, 15)
decay_rate_cands = [0.01, 0.03, 0.05, 0.1, 0.2]
plt.figure(figsize=(12, 4))
for decay_rate in decay_rate_cands:
    plt.plot(days, time_decay_weight(days, decay_rate), label=decay_rate)
plt.xlabel("elapsed_day")
plt.ylabel("weight")
plt.legend()
plt.tight_layout()
plt.show()
```
